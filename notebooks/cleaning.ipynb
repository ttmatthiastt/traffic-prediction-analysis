# Cleaning the 2‑month traffic dataset

In this notebook I:

1. Load the raw 2‑month traffic data from `../data/raw/TrafficTwoMonth.csv`.
2. Inspect the schema and basic data quality.
3. Construct a `datetime` column from `Date` and `Time`.
4. Sort the data chronologically and remove duplicates.
5. Save the cleaned dataset as `../data/clean/traffic_2months_clean.csv`.

This cleaned file is the starting point for feature engineering and forecasting.

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_rows", 10)
pd.set_option("display.max_columns", None)
```

1. Load raw data

```python
raw_path = "../data/raw/TrafficTwoMonth.csv"  
df_raw = pd.read_csv(raw_path)
df_raw.head()
```

```python
df_raw.info()
```

```python
df_raw.isna().sum()
```

### Brief interpretation of the raw schema

From the output above:

- There are 5952 rows and 9 columns.
- `CarCount`, `BikeCount`, `BusCount`, `TruckCount`, and `Total` are already `int64`.
- `Time`, `Date`, `Day of the week`, and `Traffic Situation` are objects
  (string‑like) as expected.

Since the counts are already integers and there are no missing values,
I do not need to convert types or impute data here

## 2. Copy to a working DataFrame

To avoid accidentally modifying the original object, I work on a copy
named `df`.

```python
df = df_raw.copy()
```

Some small sanity checks:

```python
numeric_cols = ["CarCount", "BikeCount", "BusCount", "TruckCount", "Total"]

(df[numeric_cols] < 0).sum() # wanted to check that the numeric columns are always non‑negative.
```

```python
component_sum = df["CarCount"] + df["BikeCount"] + df["BusCount"] + df["TruckCount"]
diff = df["Total"] - component_sum

diff.describe() #wanted to check that the `Total` column is roughly consistent with the sum of the separate vehicle counts
   
```

At this point, the "cleaning" process mainly documents that:

- There are no missing values.
- Numeric types are correct.
- Basic sanity checks have been performed.

I will save the result to `../data/clean/traffic_2months_clean.csv`.

```python
clean_path = "../data/processed/traffic_2months_clean.csv"

df.to_csv(clean_path, index=False)

clean_path
```
